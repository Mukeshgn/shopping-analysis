```python
import pandas as pd
```

```python
df =pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount_usd"})
```

```python
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult','Adult',"Middle-aged","Senior"]
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# Create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14, 'Weekly':7, 'Annually':30, 'Quarterly':90, 'Bi-Weekly':14,
       'Monthly':30, 'Every 3 Months':90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head()
```

```python
df.head()
```
